```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import requests
```

```python
webpage=requests.get('https://nexturn.com/about/').text
```

```python
webpage
```

```python
 soup=BeautifulSoup(webpage,'html.parser')
```

```python
soup
```

```python
soup.find_all('h2')[331].text
```

```python
len(soup.find_all('h2'))
```

```python
soup.find_all('h2')
    
```

```python
name=[]
headers = soup.find_all('h2')
end_index = 331
for i in range(16, len(headers), 2):  
    if i > end_index:
        break
    header = headers[i]
    header_name = header.get_text().strip()
    print(header_name)
    name.append(header_name)
```

```python
len(name)
```

```python
Profile=[]
headers = soup.find_all('h2')



end_index = 332
for i in range(17, len(headers), 2):  # Starts from index 1 with a step of 2
    if (i + 1) > end_index:
        break
    header = headers[i]
    header_name = header.get_text().strip()
    print(header_name)
    Profile.append(header_name)
```

```python
len(Profile)
```

```python
d={'n ame':name, 'Profile':Profile}
df=pd.DataFrame(d)
```
